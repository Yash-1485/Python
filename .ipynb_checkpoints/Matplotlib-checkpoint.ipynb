```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
# x,y,points
x=np.linspace(1,10,100)
plt.plot(x,np.sin(x),label = "Sine Curve")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("A sine curve")
plt.legend()
plt.grid()
plt.show()
```

```python
class Matrix:
    def __init__(self,row,col):
        self.m=[]
        self.row=row
        self.col=col
        for i in range(self.row):
            w=[]
            for j in range(self.col):
                w.append(int(input('Enter element: ')))
            self.m.append(w)
    
    def display(self):
        for i in range(self.row):
            for j in range(self.col):
                print(self.m[i][j],end=" ")
            print()
    
    def __add__(self,other):
        if self.row==other.row and self.col==other.col:
            sum_=[]
            for i in range(self.row):
                row=[]
                for j in range(self.col):
                    row.append(self.m[i][j]+other.m[i][j])
                sum_.append(row)
            return sum_
        else:return None
    
    def __mul__(self,other):
        a=self.m
        b=other.m
        row1=self.row
        col1=self.col
        row2=other.row
        col2=other.col
        if(col1==row2):
            mat=[]
            for i in range(row1):
                w=[]
                for j in range(col2):
                    ans=0
                    for k in range(col1):
                        ans+=(a[i][k]*b[k][j])
                    w.append(ans)
                mat.append(w)
            return mat
        else:
            print('Not Possible!')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
workout_durations_data = [40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55]
calories_burned_data = [200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270]

# Function to calculate weekly averages
def calculate_weekly_averages(data, days_per_week=7):
    num_weeks = len(data) // days_per_week
    weekly_averages = []
    
    for i in range(num_weeks):
        week_data = data[i * days_per_week : (i + 1) * days_per_week]
        weekly_averages.append(np.mean(week_data))
    
    return weekly_averages

# Calculate weekly averages for workout durations and calories burned
weekly_workout_averages = calculate_weekly_averages(workout_durations_data)
weekly_calories_averages = calculate_weekly_averages(calories_burned_data)

# Calculate rate of change in workout durations
def calculate_rate_of_change(data):
    return np.diff(data)

rate_of_change = calculate_rate_of_change(workout_durations_data)

# Plotting the visualizations
fig, axs = plt.subplots(3, 2, figsize=(14, 12))

# Subplot 1: Line plot for daily workout durations
axs[0, 0].plot(workout_durations_data, marker='o', markersize=8, color='blue')
axs[0, 0].set_title('Daily Workout Durations')
axs[0, 0].set_xlabel('Days')
axs[0, 0].set_ylabel('Duration (minutes)')

# Subplot 2: Line plot for weekly averages (workout duration & calories burned)
weeks = range(1, len(weekly_workout_averages) + 1)
axs[0, 1].plot(weeks, weekly_workout_averages, marker='o', markersize=8, color='green', label='Workout Duration (min)')
axs[0, 1].plot(weeks, weekly_calories_averages, marker='s', markersize=8, color='orange', label='Calories Burned')
axs[0, 1].set_title('Weekly Averages')
axs[0, 1].set_xlabel('Weeks')
axs[0, 1].set_ylabel('Average Values')
axs[0, 1].legend()

# Subplot 3: Bar chart for daily workout durations
axs[1, 0].bar(range(1, len(workout_durations_data) + 1), workout_durations_data, color='blue')
axs[1, 0].set_title('Daily Workout Durations (Bar Chart)')
axs[1, 0].set_xlabel('Days')
axs[1, 0].set_ylabel('Duration (minutes)')

# Subplot 4: Histogram for workout durations distribution
axs[1, 1].hist(workout_durations_data, bins=10, color='skyblue')
axs[1, 1].set_title('Workout Durations Distribution (Histogram)')
axs[1, 1].set_xlabel('Duration (minutes)')
axs[1, 1].set_ylabel('Frequency')

# Subplot 5: Pie chart for calories burned distribution over the weeks
weekly_calories_sum = [np.sum(calories_burned_data[i * 7:(i + 1) * 7]) for i in range(len(weekly_workout_averages))]
axs[2, 0].pie(weekly_calories_sum, labels=[f'Week {i + 1}' for i in range(len(weekly_calories_sum))],
              autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axs[2, 0].set_title('Calories Burned Distribution (Pie Chart)')

# Subplot 6: Line plot for the rate of change in workout durations
axs[2, 1].plot(range(2, len(workout_durations_data) + 1), rate_of_change, marker='^', markersize=8, color='red')
axs[2, 1].set_title('Rate of Change in Workout Durations')
axs[2, 1].set_xlabel('Days')
axs[2, 1].set_ylabel('Rate of Change (minutes/day)')

# Set the main title
fig.suptitle('Fitness Data Analysis', fontsize=16)

# Adjust the layout to prevent overlap
plt.tight_layout()
plt.subplots_adjust(top=0.92)

# Show the plot
plt.show()
```
